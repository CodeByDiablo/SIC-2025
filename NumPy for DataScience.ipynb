### Problem 1: Find the memory size of any array.



```
import numpy as np
arr = np.array([[1, 2, 3], [4, 5, 6]])

print(f"{arr}")
print(f"Number of elements in the array (size): {arr.size}")
print(f"Size of one element in bytes (itemsize): {arr.itemsize}")
print(f"Total bytes consumed by the array data (nbytes): {arr.nbytes}")
```

### Problem 8: Create a 2D array with 1 on the border and 0 inside.

```
import numpy as np
n_rows, n_cols = 5, 5
border_array = np.ones((n_rows, n_cols))
if n_rows > 2 and n_cols > 2:
    border_array[1:-1, 1:-1] = 0

print(f"2D array with 1 on border and 0 inside:\n{border_array}")
```

### Problem 9: How to add a border (filled with 0s) around an existing array using `np.pad()`.

```
import numpy as np
existing_array = np.array([[1, 2], [3, 4]])
bordered_array_pad = np.pad(existing_array, pad_width=1, mode='constant', constant_values=0)

print(f"Original array:\n{existing_array}")
print(f"Array with 0-border using np.pad():\n{bordered_array_pad}")
```

### Problem 10: How to add a border (filled with 0s) around an existing array using fancy indexing.

```
import numpy as np
existing_array = np.array([[1, 2], [3, 4]])

row_padding = 1
col_padding = 1
new_rows = existing_array.shape[0] + 2 * row_padding
new_cols = existing_array.shape[1] + 2 * col_padding
bordered_array_manual = np.zeros((new_rows, new_cols), dtype=existing_array.dtype)

rows_to_fill = np.arange(row_padding, new_rows - row_padding)
cols_to_fill = np.arange(col_padding, new_cols - col_padding)
bordered_array_manual[np.ix_(rows_to_fill, cols_to_fill)] = existing_array

print(f"Original array:\n{existing_array}")
print(f"Array with 0-border (using fancy indexing):\n{bordered_array_manual}")
```

### Problem 11: Create a 5x5 matrix with values 1,2,3,4 just below the diagonal.
